```python
import random
def tictactoe():
    
    def display_grid(L):
        result=['-'*13]
        body=[]
        for item in L:
            body.append('| '+item+' |')
        modified_body=[]
        if body !=[]:
            for i in range(3):
                modified_body.append(body[0:3])
                del(body[0:3])
        MB=[]
        for item in modified_body:
            MB.append(''.join(item))
        new_MB=[]
        for string in MB:
            new_MB.append(string.replace('||','|'))
        for item in new_MB:
            result.append('\n'+item+'\n'+'-'*13)
        return ''.join(result)
    
    def possible_moves(L):
        return [key for key, val in enumerate(L) if val == ' ']
    
    def is_winner(L, letter):
        return  (L[0] == L[1] == L[2] == letter) or \
                (L[3] == L[4] == L[5] == letter) or \
                (L[6] == L[7] == L[8] == letter) or \
                (L[0] == L[3] == L[6] == letter) or \
                (L[1] == L[4] == L[7] == letter) or \
                (L[2] == L[5] == L[8] == letter) or \
                (L[0] == L[4] == L[8] == letter) or \
                (L[2] == L[4] == L[6] == letter)
    
    def give_and_check_move(L):
        while True:
            try:
                hum=input('Please input a move: ')
                hum=int(hum)
                if L[int(hum)]!= ' ':
                    print('Must be in an unoccupied space: ' + str(hum) + ' is taken')
                    continue
                else:
                    L[int(hum)]='X'
                    break
            except ValueError:
                print('Must be an integer')
                continue
            except IndexError:
                print('Must be between 0-8')
                continue
    
    def AI_move(L):
        for i in possible_moves(L):
            L[i]='O'
            if is_winner(L,'O')is True:
                return i
            else:
                L[i]=' '
        for i in possible_moves(L):
            L[i]='X'
            if is_winner(L,'X')is True:
                L[i]=' '
                return i
            else:
                L[i]=' '
        result=[]
        for ch in [0,2,6,8]:
            if ch in possible_moves(L):
                result.append(ch)
        if result!=[]:
            return random.choice(result)
        if 4 in possible_moves(L):
            return 4
        return random.choice(possible_moves(L))
    
    counter=4
    L=[' ',' ',' ',' ',' ',' ',' ',' ',' ']
    print(display_grid(L))
    
    while possible_moves(L)!=[]:        
        give_and_check_move(L)        
        print(display_grid(L))
        
        if(is_winner(L, 'X'))==True:
            return 'Congratulations, you win!'
        if(is_winner(L, 'O'))==True:
            return 'Sorry, you lose!'
        
        if counter>0:
            L[AI_move(L)]='O'
            print(display_grid(L))
            counter=counter-1

        if(is_winner(L, 'X'))==True:
            return 'Congratulations, you win!'
        if(is_winner(L, 'O'))==True:
            return 'Sorry, you lose!'            
    
    if(is_winner(L, 'X'))==True:
            return 'Congratulations, you win!'
    if(is_winner(L, 'O'))==True:
        return 'Sorry, you lose!'
    return 'draw!'
tictactoe()
```
