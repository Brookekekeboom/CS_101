# Problem 1 - Mastermind

This is a code-breaking game that can be played by two players. More information can be found here: 

https://en.wikipedia.org/wiki/Mastermind_(board_game)

The codemaker sets the code. The code is an ordering of 4 pegs where each peg can be one of 6 different colors. For example, there might be the following 6 colors at play: red, yellow, green, blue, purple, and orange. So a possible code might be:

blue, blue, orange, red

The codebreaker now has the task of guessing the code in terms of the exact order and colors. Once a guess is submitted the codemaker provides feedback in the form of 0 to 4 black or white colored pegs. A black peg is placed for each peg from the codebreaker's guess which is correct in both color and position. A white peg indicates the existence of a correct colored peg placed in the wrong position.

This interaction takes place until 12 (or 10 or 8) incorrect guesses are made. If the codebreaker hasn't guessed the code by this time the codemaker wins. If the codebreaker manages to guess the code before the limit of incorrect guesses is reached the codebreaker wins.

**1**: When the game ends be sure to report the number of guesses taken.

**2**: Be sure to work this out using modular programming techniques!
